```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from utils import *

FIG_FOLDER = 'figures/outcomes_by_rank'
```

```python
data = load_to_df('data/twitter_data.json')
political_data = data[data["is_political"] == True]
```

```python
AUTHOR_EMOTIONS = ["author_happy", "author_sad", "author_anxious", "author_angry"]
READER_EMOTIONS = ["reader_happy", "reader_sad", "reader_anxious", "reader_angry"]
```

```python
AUTHOR_EMOTIONS_LABELS = ["Author Happy", "Author Sad", "Author Anxious", "Author Angry"]
READER_EMOTIONS_LABELS = ["Reader Happy", "Reader Sad", "Reader Anxious", "Reader Angry"]
ENGAGEMENT_COLOR = sns.color_palette()[0]
CHRONOLOGICAL_COLOR = "gray"
```

```python
yticks = [
    [x / 100.0 for x in range(113, 155, 12)],
    [x / 100.0 for x in range(24, 55, 8)],
    [x / 100.0 for x in range(29, 70, 12)],
    [x / 100.0 for x in range(27, 80, 14)]]
fig, axs = plt.subplots(1, 4, figsize=(34, 8))
for i, author_emo in enumerate(AUTHOR_EMOTIONS):
    axs[i].margins(x=0)
    g1 = sns.lineplot(data=data[data["timeline"] == "Engagement"], x="rank", y=author_emo, 
                style="timeline", marker="o", markersize=10, dashes=False, legend=False, color=ENGAGEMENT_COLOR, ax=axs[i])
    g1.set_xticks(range(1, 11))
    g1.set_yticks(yticks[i])
    g1.tick_params(labelsize=18)
    g1.set_xlabel(None)
    g1.set_ylabel(AUTHOR_EMOTIONS_LABELS[i], fontsize=22)
    g2 = sns.lineplot(data=data[data["timeline"] == "Chronological"], x="rank", y=author_emo,
                style="timeline", marker="o", markersize=10, dashes=False, legend=False, color=CHRONOLOGICAL_COLOR, ax=axs[i])
    g2.set_xticks(range(1, 11))
    g2.set_yticks(yticks[i])
    g2.tick_params(labelsize=18)
    g2.set_xlabel(None)
    g2.set_ylabel(AUTHOR_EMOTIONS_LABELS[i], fontsize=22)
engagement_patch = mpatches.Patch(color=ENGAGEMENT_COLOR, label="Engagement")
chronological_patch = mpatches.Patch(color=CHRONOLOGICAL_COLOR, label="Chronological")
g2.legend(handles=[engagement_patch, chronological_patch], loc='upper right', prop={'size': 14})
fig.suptitle("Author Emotions by Tweet Rank", fontsize=25, fontweight="bold")
fig.text(0.5, 0.0, "Rank of Tweet in Timeline", ha='center', fontsize=26)
plt.savefig(f"{FIG_FOLDER}/author_emotions.pdf");
```
